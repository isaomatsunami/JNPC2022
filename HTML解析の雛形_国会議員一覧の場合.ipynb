コピペでタグの属性データを取得することは難しい。

ファイルを読み込む。

```python
targetFileName = "raw/kokkaigiin_ichiran_list.html"

f = open(targetFileName, "r")
print( f.read() )
```

HTML parserは、HTMLの構造を分析するライブラリ。

```python
from bs4 import BeautifulSoup  # 超有名なhtmlパーサーを使う宣言
import html  # htmlモジュールのunescapeは文字化けさせないおまじない

f = open(targetFileName, "r")
bsObj = BeautifulSoup( f.read(), "html")
print( bsObj.prettify() )  # 枝構造が明瞭になるように出力する
```

このテーブルに関する以下の分析結果を利用する。

1. テーブルの各行(tr)が各議員を表す
2. 各行の２番目の列(td)は選挙区を表す
3. 各行の５番目の列(td)の、aタグは氏名を表す
4. 各行の13番目の列の、twittersクラスのaタグの属性hrefがTwitterのリンクを表す

```python
f = open(targetFileName, "r")
bsObj = BeautifulSoup( f.read(), "html")

for i, row in enumerate( bsObj.find_all("tr") ):
    if i == 0:
        continue  # 一行目は項目名なので無視
    senkyoku = row.find_all("td")[1].get_text()
    name = row.find_all("td")[4].find("a").get_text()
    twitter_a = row.find_all("td")[12].find("a", class_="twitters")
    if twitter_a:  # 空欄の場合に備えて、タグがある場合だけhrefを取得する
        twitter = twitter_a["href"]
    else:
        twitter = None
    print( senkyoku, name, twitter )
```

printの代わりにファイルに書き出す

```python
f = open(targetFileName, "r")
bsObj = BeautifulSoup( f.read(), "html")

output = open("twitter_accounts.csv","w")

for i, row in enumerate( bsObj.find_all("tr") ):
    if i == 0:
        continue  # 一行目は項目名なので無視
    senkyoku = row.find_all("td")[1].get_text()
    name = row.find_all("td")[4].find("a").get_text()
    twitter_a = row.find_all("td")[12].find("a", class_="twitters")
    if twitter_a:  # 空欄の場合に備えて、タグがある場合だけhrefを取得する
        twitter = twitter_a["href"]
    else:
        twitter = "empty"
    # 出力。項目の中にカンマがあるとcsvが壊れるので、削除する
    output.write( "{0:s},{1:s},{2:s}\n".format(senkyoku.replace(",",""), name.replace(",",""), twitter) )
```
